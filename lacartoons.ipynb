```python
import time
import requests
import json
import re
from pyquery import PyQuery as pq
```

```python
URL_TOTAL = "https://www.lacartoons.com/serie/18"
URL_BASE = "https://www.lacartoons.com"

HEADERS = {'User-Agent': "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) "
                         "AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50"
                         ".0.2661.102 Safari/537.36"}

main_page_response = requests.get(URL_TOTAL, headers=HEADERS)
main_page_response
data = []
```

```python
show_name = pq(main_page_response.content).find('h2')[0].text_content().split('\n')[0]
slug = show_name.lower().replace(' ', '_')
```

```python
for index in range(278, 400):
    url = f"https://www.lacartoons.com/serie/{index}"
    main_page_response = requests.get(url, headers=HEADERS)
    data = []
    show_name = pq(main_page_response.content).find('h2')[0].text_content().split('\n')[0]
    slug = show_name.lower().replace(' ', '_')
    seasons = pq(main_page_response.content).find('.listas-de-episodion')
    print('----> start',url, show_name, '[slug]:', slug, len(seasons))
    
    for index_season, season in enumerate(seasons):
        episodes = pq(season).find('li a')
        for index_episode, episode in enumerate(episodes):
            url = f"{URL_BASE}{episode.attrib['href']}"
            episode_response = requests.get(url, headers=HEADERS)
            sources = [pq(episode_response.content).find('iframe')[0].attrib['src']]
            name = episode.text_content().split('-')[-1].replace('\n', '').lstrip()
            d = {
                'title': name,
                'season': index_season + 1,
                'chapter': index_episode + 1,
                'description': '',
                'sources': sources
            }
            data.append(d)
            print(d)
    print('------> end', show_name)
    save_json(data, slug)
```
