## Leer lista de episodios y seasons

```python
show_data = []
```

```python
def read_episodes_seasons(link):
    show_container = requests.get(link['href'], headers=HEADERS)
    links_episodes = pq(show_container.content).find('#seasons .episodios li')
    show_data = []
    for index, link_episode in enumerate(links_episodes):
        print(f'[{index}]: ${link_episode.text_content()}')
        if link_episode.attrib['class'] != 'none' and 'aun no disponible' not in link_episode.text_content():
            season = link_episode.getchildren()[1].text_content().split(' - ')[0]
            chapter_number = link_episode.getchildren()[1].text_content().split(' - ')[1]
            title = link_episode.getchildren()[2].find('a').text_content()
            href = link_episode.getchildren()[2].find('a').attrib['href']
            d = {
                'title': title,
                'season': season,
                'chapter': chapter_number,
                'description': '',
                'href': href,
                'sources': get_sources(href)
            }
            show_data.append(d)
        else:
            print('No DATA')
    return show_data
```

```python
def get_sources(episode_link):
    episode_content = requests.get(episode_link, headers=HEADERS)
    data_post = pq(episode_content.content).find('.dooplay_player_option')[0].attrib['data-post']
    data_type = pq(episode_content.content).find('.dooplay_player_option')[0].attrib['data-type']
    data_nume = pq(episode_content.content).find('.dooplay_player_option')[0].attrib['data-nume']
    payload = {
        'action': 'doo_player_ajax',
        'post': data_post,
        'nume': data_nume,
        'type': data_type
    }
    response = requests.post(URL_POST, payload, headers=HEADER_POST)
    embed_url = json.loads(response.content)['embed_url']
    print(response.content)
    src = re.search('(?<=src=").*?(?=[\?"])', embed_url)[0]
    print(f'get sources {src}')
    return [src, ]
```

```python
def save_json(data, name):
    archivo = open(f'./shows/{name}.json', 'w')
    archivo.write(json.dumps(data))
    archivo.close()
    print(f'saved: {name}')
```
