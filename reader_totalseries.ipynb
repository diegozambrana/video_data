```python
for index, link in enumerate(content):
    print(index, link.text_content())
    chapter_response = requests.get(link.attrib['href'], headers=HEADERS)
    title = pq(chapter_response.content).find('.entry-title')[0].text_content()
    if ' – ' in title:
        title = title.split(' – ')[1]
    description = pq(chapter_response.content).find('.item_description')
    if len(description) > 0:
        description = description[0].text_content()
    else:
        description = pq(chapter_response.content).find('.entry-content p')[0].text_content()
    iframe_url = pq(chapter_response.content).find('iframe')[0].attrib['src']
    sources = get_iframe_data(iframe_url, link.attrib['href'])
    d = {
        'title': title,
        'season': int(link.text_content().split(' ')[0].split('×')[0]) or 'none',
        'chapter': int(link.text_content().split(' ')[0].split('×')[1]) or 'none',
        'description': description,
        'sources': sources,
    }
    print(d['title'], 'readed and saved')
    data.append(d)
print(len(data), 'capitulos guardados')
```

```python
def get_iframe_data(url, referer):
    headers = {
        'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
        'accept-encoding': 'gzip, deflate, br',
        'accept-language': 'es-ES,es;q=0.9,en;q=0.8',
        'cookie': 'PHPSESSID=5pabu21l3bo0u3lsljtu2lh7g6',
        'referer': referer,
        'sec-fetch-dest': 'iframe',
        'sec-fetch-mode': 'navigate',
        'sec-fetch-site': 'same-site',
        'upgrade-insecure-requests': '1',
        'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 11_1_0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36',
    }
    test_content = requests.get(url, headers=headers)
    elements = pq(test_content.content).find('.option.servers')
    sources = list(map(lambda e: get_source_from_element(url, e), elements))
    return sources
url = "https://contenedor.totalseries.org/embed/1996"
referer = 'https://totalseries.org/2020/11/daria-1x01-estimadores.html'
get_iframe_data(url, referer)
```
